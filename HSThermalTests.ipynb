```python
import pandas as pd
import math
import matplotlib.pyplot as plt

df13 = pd.read_json('Serenity13fillers.json')
dfo13 = df13[["off","on","X10","X115","X1255","X14095","X165535","X00","X015","X0255","X04095","X065535"]].T
dfo13
```

```python
df16 = pd.read_json('Serenity16fillers.json')
dfo16 = df16[["off","on","X10","X115","X1255","X14095","X165535","X00","X015","X0255","X04095","X065535"]].T
dfo16
```

```python
dfo16["PSU_delta"]=dfo16["PSU_power"]-dfo16["PSU_power"][0]
dfo16["X_power"]=dfo16["X0_power"]+dfo16["X1_power"]
dfo16["DCDC_lost_ratio"]=dfo16["PSU_delta"]/dfo16["X_power"]
dfo16.plot.scatter(x="PSU_delta",y="X_power")
print("Average PSU_delta/X_power",dfo16.DCDC_lost_ratio[1:].mean())
print("Sigma PSU_delta/X_power",math.sqrt(dfo16.DCDC_lost_ratio[1:].var()))

dfo16[["PSU_delta","X_power","DCDC_lost_ratio"]]
```
