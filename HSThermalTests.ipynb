```python
df13alIPMC = pd.read_json('Serenity13IPMC.json')
dfo13alIPMC = df13alIPMC[["off","on","X10","X115","X1255","X14095","X165535","X00","X015","X0255","X04095","X065535"]].T
dfo13alIPMC
```

```python
dfo13alIPMC["PSU_delta"]=dfo13alIPMC["PSU_power"]-dfo13alIPMC["PSU_power"][0]
dfo13alIPMC["PIM400_delta"]=(dfo13alIPMC["PIM400_Iout"]-dfo13alIPMC["PIM400_Iout"][0])*dfo13alIPMC["PIM400_Vin"]
dfo13alIPMC["X_power"]=dfo13alIPMC["X0_power"]+dfo13alIPMC["X1_power"]
dfo13alIPMC["PIM400_delta*eff"]=dfo13alIPMC["PIM400_delta"]*0.95*0.9
dfo13alIPMC["PIM2DCDC_eff"]=dfo13alIPMC["X_power"]/dfo13alIPMC["PIM400_delta"]
dfo13alIPMC["PSU2PIM_eff"]=dfo13alIPMC["PIM400_delta"]/dfo13alIPMC["PSU_delta"]
dfo13alIPMC["DCDC_lost_ratio"]=dfo13alIPMC["X_power"]/dfo13alIPMC["PSU_delta"]
dfo13alIPMC["Abs_missing_power"]=dfo13alIPMC["PIM400_delta*eff"]- dfo13alIPMC["X_power"]
dfo13alIPMC.plot.scatter(x="PIM400_delta",y="X_power")
print("Average PSU_delta/X_power",dfo13alIPMC.DCDC_lost_ratio[1:].mean())
print("Sigma PSU_delta/X_power",math.sqrt(dfo13alIPMC.DCDC_lost_ratio[1:].var()))
dfo13alIPMC.plot.scatter(x="X_power",y="PIM2DCDC_eff")

dfo13alIPMC[["PSU_volt","PIM400_Vin","PSU_delta","PIM400_temp","PIM400_delta","X_power","PSU2PIM_eff","PIM2DCDC_eff"]]
```
